```python
# non keyword argunments

def add_multiple_value(*args):
    print(type(args))
    sum = 0
    for val in args:
        sum+=val
    print(sum)
```

```python
add_multiple_value(1,2,3,4,5)
```

```python
def greet_names(greeting, *args):
    for i in args:
        print(greeting , i)
    print(len(args))
    print(args[0])
    print(args[1])
    print(args[2])
```

```python
greet_names("hello how are you !!", "krish","mohit","mohammad")
```

```python
# keyword argunments

def add(**kwargs):
    print(type(kwargs))
    print(kwargs)
    print(kwargs["a"] + kwargs["c"] + kwargs["b"])
    print(kwargs.get("c") + kwargs.get("b") + kwargs.get("d"))

add(a=2,b=3,c=4,d=5,e=6)
```

```python
def add(*args, **kwargs):
    print(type(args))
    print(type(kwargs))
    print(args)
    print(kwargs)

add(1,2,c=4,d=5,e=6)
```
