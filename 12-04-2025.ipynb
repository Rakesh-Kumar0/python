```python
"""The most "Pythonic" way to access both the index and value simultaneously while looping through a list or other 
iterable is by using the built-in enumerate() function"""

my_list = ['apple', 'banana', 'cherry']

for idx, val in enumerate(my_list):
    print(f"Index: {idx}, Value: {val}")
```

```python
for idx, val in enumerate(my_list, start=1):
    print(f"Index: {idx}, Value: {val}")
```

```python
# this code will remove duplicate from list using set

a = [1,2,3,445455,33,33,33,453,0,0,0,0,"add"]

# method 1
a = set(a)
a = list(a)
print(a)

# method 2
a = list(set(a))
print(a)
```
