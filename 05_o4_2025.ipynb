```python
a = 2
b = 3
c = 10
d = 20
```

## control flow

```python
name = "rakesh"

if name == "rakesh":
    print("Welcome instructor")
else :
    print("Welcome attendee")
```

```python
operation_name = "divide"
print("before control flow")
if operation_name  == "add" :
    print("We are adding two number")
elif operation_name == "substract":
    print("We are substracting two number")
else :
    print("We are doing nothing")

print("after control flow")
```
