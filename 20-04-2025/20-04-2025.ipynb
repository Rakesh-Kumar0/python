# clean file read

```python
import os
```

```python
file_name = "test.txt"
cwd = os.getcwd()
full_file_path = os.path.join(cwd, file_name)
print(full_file_path)
```

```python
f = open(full_file_path)
```

```python
all_contents = f.read()
```

```python
print(all_contents)
```

```python
all_contents.split("\n")
```

```python
f = open(full_file_path)
lines = f.readline()
print(lines)

lines = f.readline()
print(lines)
f.close()
```

```python
f  = open(full_file_path)
lines = f.readlines()
print(lines)
f.close()
```

```python
# with -> context manager --> handles closing of your file

with open(full_file_path) as f:
    lines = f.readlines()
    print(lines)
```

```python
with open(full_file_path, "r") as f:
    lines = f.readlines()
    print(lines)
```

```python
data_of_students = {}
```

```python
def populate_data(data):
    print(data)
    data = data.split(",")
    print(data)
    data_of_students[str(data[0])] = {}
    data_of_students[str(data[0])][str("name")] = data[1]
    data_of_students[str(data[0])][str("age")] = data[2]
```

```python
with open("./test_dummy_1.txt", "r+") as f:
    content = f.readlines()
    print(content)
    for c in content:
        populate_data(c)
```
