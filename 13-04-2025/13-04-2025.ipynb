## pattern printing

```python
for i in range(1,6) :
    print("*"*i)
```

```python
for i in range(1,6,2) :
    print("*"*i)
```

```python
rows = 3
for i in range(1, rows+1):
    dashes = "_" * (rows-i)
    stars = "*" * i
    print(dashes + stars)
```

```python
max = 3
for i in range(2,-1,-1):
    print(i*"_", end="")
    print((max-i)*"*")
```

```python
for i in range(1,4):
    print("_" * (3-i) + "*" * i)
```

```python
for i in range(2, -1, -1):
    print("_"*i,end="")
    for j in range(i, 3):
        print("*",end="")
    print()
```
