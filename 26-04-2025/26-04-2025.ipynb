```python
try :
    a = 10
    a = int(a)
    b = "1"
    print(10/a)
    int(b)
except Exception as e:
    print("We have encountered error : ", e)

# else block only run when there is no error in try block
else :
    print("No error encountered")
finally :
    print("input section executed successfully")
```
